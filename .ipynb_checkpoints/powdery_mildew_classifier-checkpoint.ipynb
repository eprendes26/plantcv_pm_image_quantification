# DEMONSTRATIVE NOTEBOOK FOR THE USE OF PLANTCV AND PYTHON FOR FEATURE-RICH QUANTITATIVE ANALYSIS IN WHEAT POWDERY MILDIEW 

This notebook is a demonstrative way to visualize a set of tools and a workflow that could be used for the quantitative analysis of leaf section images of wheat in inoculation essays with powdery mildiew. It is intended to show the logic behind the functions and approaches utilized.

It's remotely running on a [GitHub](https://github.com/eprendes26/plantcv_pm_image_quantification) repository thanks to [Binder](https://mybinder.org/). In that same repo there's a set of seven example images (with names ranging from `example_1.jpg` to `example_7.jpg`) The examples 1 to 3 are infected plants, while examples 4 to 6 are healthy plants and example 7 is an image with a flare/artifact on it. To run any of the examples provided just write the filename of the example you want to use (`example_1.jpg`, `example_2.jpg`, etc.) in parameter of the `pcv.readimage` function in the **Image Upload** box. 

Since the classification/segmentation part of this workflow is done using the *Naive Bayes Classifier* tool that **PlantCV** has, the first step for using the classifier, which is the obtention of the *Probability Density Functions* (PDFs) from a training set, is not in the scope of this notebook, but it is a pre-requisite to have the `naive_bayes_pdfs.txt` file with the PDFs in order to be able to performs the analysis shown here. In the [**PlantCV** documentation](https://plantcv.readthedocs.io/en/latest/machine_learning_tutorial/) there is a detailed tutorial on how to proceed to obtain this PDFs from a sample of pixels.

The PDFs file used here was obtained after using a training set of 20 randomly selected images from the original database and manually sampling pixels that belongued to each one of the predefined categories/classes with the aid of the **Pixel Inspection Tool** in [ImageJ](https://imagej.nih.gov/ij/). This categories/classes were *Background*, *Infected* and *Healthy Tissue*, and 180 pixels per each were sampled. 

You can perform this analysis on your own set of images by writing the local path to the image you want to analyze as the parameter of the `pcv.readimage` function, and by using your own trained classifier (PDFs file) by writing the local path to it in the `pdf_file` argument of the `pcv.naive_bayes_classifier` function in the code down below.

At the end of the notebook a "for" loop can be found which can copied and pasted in a separate script an executed in a folder containing a set of images in order to obtain the optimized measurements and metrics described here.

### Environment setup
We import the libraries needed for the analysis and set the debug parameters to *plot* so the output of the code is visible in this notebook

```python
#Packages needed for the analysis
import os
import cv2
import numpy as np
from plantcv import plantcv as pcv
from PIL import Image

#Debug parameters setting
pcv.params.debug = "plot"
```

# Image Upload

The images are in the same folder as this script/notebook. Any of the seven images located in the [GitHub](https://github.com/eprendes26/plantcv_pm_image_quantification) repo can be used, just change the name between the *()* in the code below. As explained above, you can use your own image if you write the local path to that image in that same space

```python
# Read in a color image
img, path, filename = pcv.readimage("example_3.jpg")
```

# Pixel classification (Segmentation)

We use the Probability Density Functions (PDFs) file to classify the pixels of the image according to 3 classes we previously defined. The classes are, **Normal plant tissue**, **Background** and **Infection**. In the images below we can see (in white) the pixels that represent each class, in that very same order. Each category is now a mask that we could use for further analysis.

```python
# Classify the pixels using the pdfs file previously created and create a mask for each category
masks = pcv.naive_bayes_classifier(img, pdf_file="naive_bayes_pdfs.txt")
```

### False color image of the three classes

For easier visualization we plotted a false color image with a color for each of the classes, being <span style="color:gray">**Gray**</span> = **Background**, <span style="color:red">**Red**</span> = **Infection** and <span style="color:green">**Green**</span> = **Leaf Tissue**


```python
# Plot an image representing the classification
plotted_image = pcv.visualize.colorize_masks(masks=[masks['plant'], masks['infection'], masks['background']], 
                                       colors=['green', 'red', 'gray'])
```

**As you can see there's some noise in these results, having a lot of missclassified pixels (false positives mostly) in the edges of the leaf cuttings and in some of the leaf striations.** 

A way to see how serious is this is by using a completely healthy plant from the beginning and using the following code for measuring the porcentage of infected area relative to the total plant area. Feel free to start the analysis from the beginning using a healthy image (from example 4 to 6) to see it.

```python
#Calculation of the porcentage of disease plant given the whole plant
whole_leaf_area = np.count_nonzero(masks['plant']) + np.count_nonzero(masks['infection']) 
disease_percentage = np.count_nonzero(masks['infection']) /whole_leaf_area
print("Ratio Infected Tissue/Healthy Tissue:")
print(disease_percentage)                                          
```

If a healthy leaf section was used for the previous analysis the value will generally be around 1-5%, which is not a bad error margin, but we could still optimize it a bit more.

# Classification optimization with denoising

We can use some denoising in order to reduce the amount of missclassified pixels and to obtain a more optimized classification. In this case we use the *fill* function to eliminate small clusters of wrongly classified pixels in the **Background** and **Infection** masks and then we merge the newly denoised masks and invert them to have a filled third mask that correspond to the **Healthy plant tissue** (4th image down below)

We used *size=95* in the filling of the **Infection** mask because we concluded that most of the missclasiffied pixel clusters where smaller than that, while the *size=10000* in the filling of the **Background** mask was chosen because in the background image artifacts and missclassified clusters could be very large.

```python
#We fill the infection mask and the background one
new_infection = pcv.fill(bin_img=masks['infection'], size=95)
new_background = pcv.fill(bin_img=masks['background'], size=10000)
#Now we join those two binary images we just created and we invert them to have a new plant mask
joint_masks = pcv.logical_or(bin_img1=new_infection, bin_img2=new_background)
new_plant = pcv.invert(joint_masks)
```

### False color image of the new three classes (after denoising)

To visualize the results we create false color images with the representation of the classes before and after the denoising

```python
# Plot an image representing the classification before and after the denoising
pcv.plot_image(plotted_image)
plotted_image_denoising = pcv.visualize.colorize_masks(masks=[new_plant, new_infection, new_background], 
                                       colors=['green', 'red', 'gray'])
```

It is obvious the second image (denoised) have a lot less of noise and missclassified pixels than the one without the denoising. There are still some false positives, specially near the edges of the leaf cutting but less than in the previous one. A way of quantifying the difference is to calculate the infection percentage once again, but with this new masks

```python
#Calculation of the porcentage of disease plant given the whole plant
whole_leaf_area = np.count_nonzero(new_plant) + np.count_nonzero(new_infection) 
disease_percentage_denoised = np.count_nonzero(new_infection) /whole_leaf_area
print("Ratio Infected Tissue/Healthy Tissue after denoising:")
print(disease_percentage_denoised)                                          
```

The value is gonna be smaller than the previous one that was calculated. If you want to quantify how much does this approach helps when handling false positives and artifacts in the image that could be missclassified as background use a picture with a healthy plant from the beginning of the analysis and compare.

# More optimization. Edge cropping

Now we are going to try to get rid of the edges of the leaf cutting, since its the part where we can find most of the missclassified pixels after the denoising. In order to do so the following steps are implemented on this script:

- Extraction of the original image resolution in two variables, *width* and *height*
- Rescaling of the background mask to a 104% of the original resolution using the `transform.resize`function in **PlantCV** and the previously created variables without using interpolation, effectively increasing the size of the background and reducing that of the plant (zero-padding).
- Rescaling what was obtain in the previous step back to the original resolution using interpolation, so the plant keeps being smaller than the original (resizing)
- Use the new rescaled mask and superpose it with the plant and the infection to create a new image and infection masks without the borders

```python
#Reading the image resolution and calculating the width and height we will use for the reescaling
image = Image.open(filename)
width, height = image.size
new_width = width + (width * 0.04)
new_height = height + (width * 0.04)
```

### Obtention of the rescaled background mask (4rd image)

```python
#Inverting the background mask, rescaling it to be 4% smaller, and then re-inverting it and upscaling it back to the 
#original resolution
new_background_inverted = pcv.invert(new_background)
background_resized = pcv.transform.resize(new_background_inverted, size=(int(new_width), int(new_height)), interpolation=None)
background_resized = pcv.invert(background_resized)
background_resized = pcv.transform.resize(background_resized, size=(int(width), int(height)), interpolation="auto")
```

### Masks joining and obtention of the rescaled infection and plant mask (2nd and 3rd images)

```python
#Joining with an AND operator the infection mask and the inverted resized background mask
background_resized_inverted = pcv.invert(background_resized)
infection_resized = pcv.logical_and(background_resized_inverted, new_infection)
#Doing the same but with the healthy plant mask
plant_resized = pcv.logical_and(background_resized_inverted, new_plant)
```

### False color image of the new three classes (after edge cropping)

To visualize the results we create false color images with the representation of the classes before and after the denoising, and after the edge cropping.

```python
## Plot an image representing the classification before and after the denoising, and after the leaf edge elimination
pcv.plot_image(plotted_image)
pcv.plot_image(plotted_image_denoising)
plotted_image_denoising = pcv.visualize.colorize_masks(masks=[plant_resized, infection_resized, background_resized], 
                                       colors=['green', 'red', 'gray'])
```

When we calculated the *infected area/whole leaf* area ratio we obtain a smaller value. If we use a completely healthy plant example, we'll see that the value is going to be much closer to the theorethical real value of 0

```python
#Calculation of the porcentage of disease plant given the whole plant
whole_leaf_area = np.count_nonzero(plant_resized) + np.count_nonzero(infection_resized) 
disease_percentage_no_edge = np.count_nonzero(infection_resized) /whole_leaf_area
print("Ratio Infected Tissue/Healthy Tissue after edge cutting:")
print(disease_percentage_no_edge)                                          
```

# Extracting more features

We finish with the most important part of our analysis, the segmentation/classification part, and by doing it we already obtained an important metric, the ratio between the *infected tissue* and the *healthy tissue* of the leaf cutting. Now lets try to obtain other metrics that could be relevant.

First thing we should try is to identify the colonies/objects and to count them. In our dataset, given the resolution of the images and the fact that many of the fungal colonies are overlapping each other this might prove a very difficult task. Nevertheless, to identify the number of this colonies or groups of colonies, and to find its average size and perimeter could still provide us with a very important information.

### Finding the edge of the fungal formations

First we use PlantCV function `cany_edge_detect` to find the edges of the formation and obtain a mask with only then (1st image). After that we use **OpenCV** `dilate` function to fill the gaps in the edge line and optimize the following steps (2nd image).

```python
#By using the canny edge function in PlantCV we obtain a mask with the contour of the fungal formations
edges = pcv.canny_edge_detect(img=infection_resized, sigma=4)
#To optimize the next step we use the dilate function to fill the gaps in the edge lines 
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(4,4))
dilated = cv2.dilate(edges, kernel)
pcv.plot_image(dilated)
```

### Using OpenCV to display the contours and count the number of objects

Now, after we obtain a mask with the filled contours we use a series of **OpenCV** function that allows us to find the contours and to store them as a list, so they can be counted and also used for other analyses.

In this image we can see the final output of this step, which is the original image with the overlapped contour count. 

```python
#Finding the contours and storing them
contours, hierarchy = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2:]
#Counting the numbers of contours found
number_of_objects_in_image= len(contours)
print ("The number of objects in this image: ", str(number_of_objects_in_image))
#Drawing the contours on the original image and numerating then
contours_leaf = cv2.drawContours(img, contours, -1, (0, 255, 0), 3)
sorted_contours= sorted(contours, key=cv2.contourArea, reverse= True)
for (i,c) in enumerate(sorted_contours):
    M= cv2.moments(c)
    cx= int(M['m10']/M['m00'])
    cy= int(M['m01']/M['m00'])
    cv2.putText(contours_leaf, text= str(i+1), org=(cx,cy),
            fontFace= cv2.FONT_HERSHEY_SIMPLEX, fontScale=1, color=(255,0,0),
            thickness=2, lineType=cv2.LINE_AA)
pcv.plot_image(contours_leaf)
```

### Extracting the features from the objects

Having the objects (contours) stored, with the help of **OpenCV** we can extract different features from these objects, like area, perimeter, shape, etc. We then find the average of each feature for the image. 



```python
summatory = 0
for index, value in enumerate(contours):
    ctr_area = cv2.contourArea(contours[index])
    summatory = summatory + ctr_area
avg_area = summatory/number_of_objects_in_image
summatory = 0
for index, value in enumerate(contours):
    ctr_perimeter = cv2.arcLength(contours[index],True)
    summatory = summatory + ctr_perimeter
avg_perimeter = summatory/number_of_objects_in_image
print("The average area of each object is: ", str(avg_area))
print("The average perimeter of each object is: ", str(avg_perimeter))
```

And now we have another two more metrics!
